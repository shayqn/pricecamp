```python
import sklearn
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
import urllib
%matplotlib inline
import urllib.request, json 
import time
from tqdm import tqdm
```

```python
#load in the data
df = pd.read_csv('/Users/shayneufeld/Dropbox/insight/pricecamp/data/sales.csv',index_col=0)
df.head(5)
```

```python
df.shape
```

```python
for i in tqdm(range(300)):
    with urllib.request.urlopen('https://bandcamp.com/api/salesfeed/1/get?start_date=1497925671') as url:
        data = json.loads(url.read().decode())
        
    events = data['events']
    for event in events:
        if event['event_type'] == 'sale':
            sale = event['items'][0]
            df = df.append(pd.DataFrame(data = sale,index=[0]))
            
    df = df.drop_duplicates()
    df.to_csv('/Users/shayneufeld/Dropbox/insight/pricecamp/data/sales.csv')
    time.sleep(120 + ((-0.5)*np.random.randn(1))*4)
```
