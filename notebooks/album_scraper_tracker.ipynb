```python
import os
from time import gmtime, strftime
import time
```

```python
track = True
old_num = len(os.listdir('/Users/shayneufeld/Dropbox/insight/pricecamp/data/albums_info/'))
while track == True:
    files = os.listdir('/Users/shayneufeld/Dropbox/insight/pricecamp/data/albums_info/')
    new_num = len(files)
    diff = new_num - old_num
    time_now = strftime("%Y-%m-%d %H:%M:%S", gmtime())
    print('Time %s: Number of new albums scraped: %.0f (total %.0f)' % (time_now,diff,new_num))
    old_num = new_num
    time.sleep(60)
```
